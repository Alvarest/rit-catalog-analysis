## Exploring RIT's Black Hole Merger Simulation Catalog
This Jupyter Notebook is a small example of how to use this project to **explore and analyze** RIT's blackhole merger simulation catalog. 

In this example, we will extract a sample of 10 precessing simulations and compute the luminosity and stokes parameter of each one. 

## Loading and filtering the data
You can load the entire catalog using the `parse_catalog` function from `rit_catalog_parser`. To avoid repeating this step, you can also save the catalog into a csv using `save_csv()`.

```python
catalog = parse_catalog(link)
print(catalog.size)
```

The catalog is stored as a **Pandas DataFrame**, so you can perform all the usual operations to it. 

Here, we filter the catalog to include only **precessing simulations**.

```python
precessing = catalog[catalog['type'] == "Precessing"]
print(precessing.size)
```

There are **13464 precessing simulations** in the catalog. Let's take a small sample of 10 simulations to perform our analysis. 

```python
sample = precessing.sample(10)
```

## Computing properties of the simulation

Now we will get the luminosities and stokes parameters of all the simulations of our sample. For doing that we can use the functions getLuminosity and getStokes from merger_analysis.

```python
for index, sim in sample.iterrows():
    lum = getLuminosity(sim)
    stokes = getStokes(sim)
    print(f'Simulation number {index} \t Luminosity: {lum:.2f} \t Stokes: {stokes:.2f}')
```
